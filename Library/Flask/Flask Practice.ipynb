```python
from flask import Flask, url_for, redirect, render_template
```

```python
app = Flask(__name__)

@app.route('/')
def home():
    return "This is Homepage!"

@app.route("/name/<name>")
def name(name):
    return f"Hello, {name}"

@app.route("/admin")
def admin():
    return redirect(url_for('name', name="Chirag"))

if __name__ == "__main__":
    app.run()

    
```

```python
app = Flask(__name__)

@app.route("/")
def home():
    return render_template('Index.html', content=['Chirag', 'Deepu'])

@app.route("/name")
def names():
    return redirect(url_for('home'))

if __name__ == "__main__":
    app.run()
```
