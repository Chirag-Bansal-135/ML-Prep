```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target_name'] = iris.target
df.head()
```

```python
df['name'] = df.target_name.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0 = df[df.target_name == 0]
df1 = df[df.target_name == 1]
df2 = df[df.target_name == 2]
```

```python
import matplotlib.pyplot as plt
plt.xlabel('length')
plt.ylabel('width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red',marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='green', marker="*")
```

```python
X = df.drop(['name', 'target_name'], axis='columns')
target = df.target_name
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, train_size=0.7)
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Practice Problem
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
X = pd.DataFrame(digits.data, columns=digits.feature_names)
X
target = digits.target
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, train_size=0.8)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_G = SVC(gamma=10)
model_G.fit(X_train, y_train)
model_G.score(X_test, y_test)
```
