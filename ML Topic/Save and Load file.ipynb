```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Practice Problem
sample_df = pd.read_csv('canada_per_capita_income.csv')
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Income')
plt.scatter(sample_df.year, sample_df['per capita income (US$)'], color='blue')
sample_new_df = sample_df.drop('per capita income (US$)', axis='columns')
income = sample_df['per capita income (US$)']
reg = linear_model.LinearRegression()
reg.fit(sample_new_df, income)
reg.predict([[2020]])
```

```python
import pickle
```

```python
with open('model-pickle','wb') as f:
    pickle.dump(reg,f)
```

```python
with open('model-pickle','rb') as f:
    mp = pickle.load(f)
```

```python
mp.predict([[2020]])
```
