```python
import pandas as pd
df = pd.read_csv('heights.csv')
df.shape
```

```python
df.describe()
```

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
df[df['height'] > max_threshold]
```

```python
min_threshold = df.height.quantile(0.05)
min_threshold
```

```python
df[df.height < min_threshold]
```

```python
df_no_outlier = df[(df.height > min_threshold) & (df.height < max_threshold)]
df_no_outlier.shape
```
