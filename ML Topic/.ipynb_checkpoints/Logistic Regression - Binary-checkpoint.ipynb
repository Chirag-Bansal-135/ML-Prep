```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("insurance_data.csv")
df
```

```python
%matplotlib inline 
plt.scatter(df.age,df.bought_insurance)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance, train_size=0.8)
X_test
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_train,y_train)
```

```python
model.predict_proba(X_test)
```

```python
# Practice Problem
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Group data by salary and calculate retention (left = 0 means stayed, left = 1 means left)
salary_retention = df.groupby('salary')['left'].value_counts(normalize=True).unstack() * 100

# Rename columns for clarity
salary_retention.columns = ['Stayed (%)', 'Left (%)']

# Plot the bar graph
salary_retention.plot(kind='bar', figsize=(8, 5))
plt.title('Impact of Salary on Employee Retention', fontsize=14, weight='bold')
plt.xlabel('Salary Level', fontsize=12)
plt.ylabel('Percentage of Employees', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Employee Status')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
ren.columns=['Satyed(%)','Left(%)']
ren
```

```python
import matplotlib.pyplot as plt

# Create a histogram of satisfaction level based on whether employee left or not
plt.hist(
    [df[df['left'] == 0]['satisfaction_level'], df[df['left'] == 1]['satisfaction_level']],
    bins=10,
    label=['Stayed', 'Left'],
    stacked=True,
    alpha=0.7
)

plt.title('Employee Satisfaction Level: Stayed vs Left', fontsize=14, weight='bold')
plt.xlabel('Satisfaction Level', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
dummies = pd.get_dummies(df.salary)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
X = merged.drop(['salary','high','Department','left','Work_accident','time_spend_company','number_project','last_evaluation'], axis='columns')
X
```

```python
y = df.left
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.predict(X_test))
y_test
```

```python
model.Score(X_train, y_train)
```
