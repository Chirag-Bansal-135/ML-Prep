```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[46:52]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, train_size=0.7)
```

```python
model = SVC(kernel='rbf', C=30, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(SVC(kernel='linear', C=10, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', C=10, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', C=20, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
kernels = ['rbf', 'linear']
C = [1,10,20]
for ker in kernels:
    for C_val in C:
        score = cross_val_score(SVC(kernel=ker, C=C_val, gamma='auto'), iris.data, iris.target, cv=5)
        avg_score = np.average(score)
        print('{}_{}: {}'.format(ker,C_val,avg_score))
```

```python
clf = GridSearchCV(SVC(gamma='auto'), {
    'C' : [1,10,20],
    'kernel' : ['rbf', 'linear']
}, cv=5, return_train_score=False)
clf.fit(iris.data, iris.target)
df = pd.DataFrame(clf.cv_results_)
```

```python
df[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

model_params = {
    'svm' : {
        'model' : SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20],
            'kernel' : ['rbf','linear']
        }
    },
    'randomforest' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [1,5,10]
        }
    },
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear', multi_class='auto'),
        'params' : {
            'C' : [1,5,10]
        }
    }
}
```

```python
scores = []
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
digits = load_digits()
dir(digits)
```

```python
models_params = {
    'svm' : {
        'model' : SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20],
            'kernel' : ['linear','rbf']
        }
    },
    'randomforest' :{
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [1,5,10]
        }
    },
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear', multi_class='auto'),
        'params' : {
            'C' : [1,5,10]
        }
    },
    'descision_tree' : {
        'model' : DecisionTreeClassifier(random_state=42),
        'params' : {
            'max_depth': [10, 20, 30, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'multinomial_NB' : {
        'model' : MultinomialNB(),
        'params' : {}
    },
    'gaussian_NB' : {
        'model' : GaussianNB(),
        'params' : {}
    }
}
```

```python
result = []
for model_name, mp in models_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(digits.data, digits.target)
    result.append({
        'model' : model_name,
        'best score' : clf.best_score_,
        'best param' : clf.best_params_
    })

df = pd.DataFrame(result, columns=['model', 'best score', 'best param'])
df
```
