```python
#Practice Problem
df = pd.read_csv('heart.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
le = LabelEncoder()
df['RestingECG'] = le.fit_transform(df['RestingECG'])
df.Sex = le.fit_transform(df.Sex)
df.ChestPainType = le.fit_transform(df.ChestPainType)
df.ExerciseAngina = le.fit_transform(df.ExerciseAngina)
df.ST_Slope = le.fit_transform(df.ST_Slope)
df
```

```python
from scipy.stats import zscore
```

```python
df['zscore_RestingBP'] = zscore(df['RestingBP'])
df['zscore_Cholesterol'] = zscore(df['Cholesterol'])
df['zscore_MaxHR'] = zscore(df['MaxHR'])
df['zscore_Oldpeak'] = zscore(df['Oldpeak'])

df = df[df['zscore_RestingBP'] < 3]
df = df[df['zscore_RestingBP'] > -3]
df = df[df['zscore_Cholesterol'] < 3]
df = df[df['zscore_Cholesterol'] > -3]
df = df[df['zscore_MaxHR'] < 3]
df = df[df['zscore_MaxHR'] > -3]
df = df[df['zscore_Oldpeak'] < 3]
df = df[df['zscore_Oldpeak'] > -3]

```

```python
ct = ColumnTransformer([('heart',OneHotEncoder(), ['RestingECG','Sex','ChestPainType','ExerciseAngina','ST_Slope'])], remainder='passthrough')
X = ct.fit_transform(df.drop(['HeartDisease'], axis='columns'))
X.shape
```

```python
y = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
model_svm = SVC().fit(X_train, y_train)
print(model_svm.score(X_test, y_test))
model_logistic_regression = LogisticRegression(max_iter=1000).fit(X_train, y_train)
print(model_logistic_regression.score(X_test, y_test))
model_random_forest = RandomForestClassifier().fit(X_train, y_train)
print(model_random_forest.score(X_test, y_test))
```

```python
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```
