```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices dummy.csv')
df
```

```python
# Using pandas to create dummy vaiable 
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# Dummy Variable Trap
# When you can derive one variable from other variables, they are known 
# to be multi-colinear. Here if you know values of california and georgia 
# then you can easily infer value of new jersey state, i.e. california=0 
# and georgia=0. There for these state variables are called to be multi-
# colinear. In this situation linear regression won't work as expected. 
# Hence you need to drop one column.

final = merged.drop(['town','monroe township'], axis='columns')
final
```

```python
X = final.drop(['price'], axis='columns')
y = final.price
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[3000,0,0]])   #house price of Monroe Township
```

```python
reg.score(X,y)
```

```python
# One Hot Encoder
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn import linear_model
```

```python
dfle = pd.read_csv('homeprices dummy.csv')
dfle
```

```python
le = LabelEncoder()
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town', 'area']].values
y = dfle.price.values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(),[0])],remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.score(X,y)
```

```python
# Practice Problem with Dummies
import pandas as pd
from sklearn import linear_model
import numpy as np
```
