```python
import math
import numpy as np
```

```python
#Activation Function
def sigmoid(x):
    return 1/(1+math.exp(-x))       #Output range: 0 to 1

def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))           #Output_range: -1 to 1

def relu(x):
    return max(0,x)

def leaky_relu(x):
    return max(0.1*x, x)
```

```python
# Mean absoulte error
y_pred = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])

def mae(y_pred, y_true):
    total_error = 0
    for yp, yt in zip(y_pred, y_true):
        total_error += abs(yp-yt)
    print("total error is: ", total_error)
    mae = total_error/len(y_pred)
    print("Mean absolute error is: ", mae)
    return mae

mae(y_pred, y_true)
```

```python
# Absoulte mean error with Numpy
def mae_np(y_pred, y_true):
    return np.mean(np.abs(y_pred - y_true))
```

```python
# Implement Log loss or Binary Cross Entropy
epsilon = 1e-15
def log_loss(y_pred, y_true):
    y_pred_new = [max(i, epsilon) for i in y_pred]
    y_pred_new = [min(i, 1-epsilon) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    log_loss = -np.mean(y_true * np.log(y_pred_new) + (1-y_true)*np.log(1-y_pred_new))
    return log_loss
```

```python
log_loss(y_pred, y_true)
```

```python
# Mean Squared Error
def MSE(y_pred, y_true):
    squared = 0
    for yp, yt in zip(y_pred, y_true):
        squared += (yp-yt)**2
    mse = squared/len(y_pred)
    return mse
print(MSE(y_pred, y_true))
```
