```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa','profile_score','lpa'])
df
```

```python
def initilaize_parameter(layer_dims):
    parameter = {}
    L = len(layer_dims)

    for l in range(1,L):
        parameter['W'+str(l)] = np.ones((layer_dims[l-1],layer_dims[l]))*0.1
        parameter['b'+str(l)] = np.zeros((layer_dims[l],1))

    return parameter

initilaize_parameter([2,2,1])
```

```python
def linear_forward(A_prev, W, b):
    Z = np.dot(W.T, A_prev) + b
    return Z
```

```python
def L_layer_forward(X, parameter):
    A = X
    L = len(parameter) // 2

    for l in range(1, L+1):
        A_prev = A
        Wl = parameter['W'+str(l)]
        bl = parameter['b'+str(l)]
        # print("A"+str(l-1)+": ", A_prev)
        # print('W'+str(l)+': ', Wl)
        # print('b'+str(l)+': ', bl)
        # print("--"*20)

        A = linear_forward(A_prev, Wl, bl)
        # print('A'+str(l)+': ', A)
        # print("**"*20)

    return A, A_prev
```

```python
X = df[['cgpa','profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]

parameter = initilaize_parameter([2,2,1])

L_layer_forward(X, parameter)
```

```python
def update_parameters(parameters, y, y_hat, A1, X):
    parameter['W2'][0][0] = parameters['W2'][0][0] + (0.001*2*(y-y_hat)*A1[0][0])
    parameter['W2'][1][0] = parameters['W2'][1][0] + (0.001*2*(y-y_hat)*A1[1][0])
    parameter['b2'][0][0] = parameters['W2'][1][0] + (0.001*2*(y-y_hat))

    parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
    parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
    parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])
    
    parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
    parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
    parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]

parameters = initilaize_parameter([2,2,1])

y_hat,A1 = L_layer_forward(X,parameters)
y_hat = y_hat[0][0]

update_parameters(parameters,y,y_hat,A1,X)

parameters
```

```python
# Epochs
paramters = initilaize_parameter([2,2,1])
epochs = 5
loss = []
for i in range(epochs):
    
    for j in range(df.shape[0]):
        X = df[['cgpa','profile_score']].values[j].reshape(2,1)
        y = df[['lpa']].values[j][0]

        y_hat, A1 = L_layer_forward(X, parameters)
        y_hat = y_hat[0][0]

        update_parameters(parameters, y, y_hat, A1, X)
        loss.append((y-y_hat)**2)
    print('Epoch - ', i+1, 'loss - ', np.array(loss).mean())
```
