```python
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(28,28,1)),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu'),

    keras.layers.Flatten(),

    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
model.summary()
```

```python
# Padding: It adds extra rows and columns in the matrix on right, left, up and down
#          and values of each of those cell is 0.
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu', input_shape=(32,32,1)),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'),

    keras.layers.Flatten(),

    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
model.summary()
```

```python
# Strides: When we convolve, we move 1 step and do convolution each time. in strides, we can take jumps.
#          If stride=2, we move sideways and down 2 step each time.
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', strides=(2,2), input_shape=(32,32,1), activation='relu'),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', strides=(2,2), activation='relu'),
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', strides=(2,2), activation='relu'),

    keras.layers.Flatten(),

    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
model.summary()
```
