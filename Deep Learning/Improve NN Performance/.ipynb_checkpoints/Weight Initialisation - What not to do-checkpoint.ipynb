```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ushape.csv')
df.columns = ['col1', 'col2', 'y']
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
import tensorflow as tf
from tensorflow import keras
```

# Case 1: When weights is initial with Zero

```python
model = keras.Sequential([
    keras.layers.Dense(10, activation='sigmoid', input_dim=2),
    keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.zeros(model.get_weights()[0].shape)
initial_weights[1] = np.zeros(model.get_weights()[1].shape)
initial_weights[2] = np.zeros(model.get_weights()[2].shape)
initial_weights[3] = np.zeros(model.get_weights()[3].shape)
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, epochs=100, validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype('int'), clf=model, legend=2)
```

### When sigmoid is used, it will return same values of simialar node weights. If we use tanh or relu activation function, it will return all weight zero

# Case 2: When weight is non zero constant

```python
model1 = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_dim=2),
    keras.layers.Dense(1, activation='relu')
])
model1.summary()
```

```python
model1.get_weights()
```

```python
initial_weights = model1.get_weights()
```

```python
initial_weights[0] = np.ones(model1.get_weights()[0].shape)*0.5 
initial_weights[1] = np.ones(model1.get_weights()[1].shape)*0.5 
initial_weights[2] = np.ones(model1.get_weights()[2].shape)*0.5 
initial_weights[3] = np.ones(model1.get_weights()[3].shape)*0.5 
```

```python
model1.set_weights(initial_weights)
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='Adam')
```

```python
history1 = model.fit(X, y, validation_split=0.2, epochs=100)
```

```python
plot_decision_regions(X, y.astype(int), clf=model1, legend=2)
plt.show()
```

```python
model.get_weights()
```

### In this case, sigmoid, relu and tanh will give same weight value of from similar outgoing nodes.

# Case 3.1: When initialisation weight is random, but weight has small values

```python
model2 = keras.Sequential([
    keras.layers.Dense(2, activation='tanh', input_dim=2),
    keras.layers.Dense(2, activation='tanh'),
    keras.layers.Dense(2, activation='tanh'),
    keras.layers.Dense(2, activation='tanh'),
    keras.layers.Dense(1, activation='sigmoid')
])
model2.summary()
```

```python
model2.get_weights()
```

```python
initial_weights = model2.get_weights()
```

```python
initial_weights[0] = np.random.randn(model2.get_weights()[0].shape[0], model2.get_weights()[0].shape[1])*0.01
initial_weights[1] = np.zeros(model2.get_weights()[1].shape)
initial_weights[2] = np.random.randn(model2.get_weights()[2].shape[0], model2.get_weights()[2].shape[1])*0.01
initial_weights[3] = np.zeros(model2.get_weights()[3].shape)
initial_weights[4] = np.random.randn(model2.get_weights()[4].shape[0], model2.get_weights()[4].shape[1])*0.01
initial_weights[5] = np.zeros(model2.get_weights()[5].shape)
initial_weights[6] = np.random.randn(model2.get_weights()[6].shape[0], model2.get_weights()[6].shape[1])*0.01
initial_weights[7] = np.zeros(model2.get_weights()[7].shape)
initial_weights[8] = np.random.randn(model2.get_weights()[8].shape[0], model2.get_weights()[8].shape[1])*0.01
initial_weights[9] = np.zeros(model2.get_weights()[9].shape)
```

```python
model2.set_weights(initial_weights)
```

```python
model2.get_weights()
```

```python
model2.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history2 = model2.fit(X, y, epochs=100, validation_split=0.2, verbose=0)
```

```python
plot_decision_regions(X, y.astype(int), clf=model2, legend=2)
```

```python
model2.get_weights()
```

### In this case, tanh and sigmoid will show vanishing gradient problem. Relu can converge but it will take a lot of time.

# Case 3.2: When initialisation weight is random, but weight has large values

### In this case, Sigmoid and tanh will give saturating (squeeze down values in range of values) values and training will be slow. In relu function, it will show zig-zag motion and will not be able to converge
