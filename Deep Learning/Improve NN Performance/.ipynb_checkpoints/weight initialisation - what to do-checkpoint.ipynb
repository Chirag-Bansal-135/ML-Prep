```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('ushape.csv')
df.columns = ['col1', 'col2', 'y']
```

```python
df.head()
```

```python
plt.scatter(df['col1'], df['col2'], c=df['y'])
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:,-1].values
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_dim=2, kernel_initializer='he_normal'),
    keras.layers.Dense(10, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dense(10, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dense(10, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
histroy = model.fit(X, y, epochs=300, validation_split=0.2, verbose=0)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype(int), clf=model, legend=2)
```

### When you create layers for the model NN, in kernel initializer used he_normal, he_uniform, glorot_uniform(default), glorot_normal. It will initialize weights with best possible values.
