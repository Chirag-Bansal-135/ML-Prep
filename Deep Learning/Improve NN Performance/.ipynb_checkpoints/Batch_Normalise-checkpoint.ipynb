```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('3.concertriccir2.csv')
df.columns = ['col1', 'col2', 'y']
df.head()
```

```python
plt.scatter(df['col1'], df['col2'], c=df['y'])
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model1 = keras.Sequential([
    keras.layers.Dense(2, activation='relu', input_dim=2),
    keras.layers.Dense(2, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model1.summary()
```

```python
model1.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
history1 = model1.fit(X, y, epochs=200, verbose=0, validation_split=0.2)
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(3, activation='relu', input_dim=2, kernel_initializer='he_normal'),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(2, activation='relu', kernel_initializer='he_normal'),
    keras.layers.BatchNormalization(),
    # keras.layers.Dense(2, activation='relu', kernel_initializer='he_normal'),
    # keras.layers.BatchNormalization(),
    keras.layers.Dense(1, activation='sigmoid')
])
model2.summary()
```

```python
model2.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
history2 = model2.fit(X, y, validation_split=0.2, verbose=0, epochs=200)
```

```python
plt.plot(history1.history['val_accuracy'], color='black')
plt.plot(history2.history['val_accuracy'], color='red')
```
